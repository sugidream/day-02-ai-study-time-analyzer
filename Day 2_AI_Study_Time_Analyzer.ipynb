```python
# 🚀 Day 02: AI Study Time Analyzer

> Turning study hours into actionable insights using Python & NumPy 📊

## 🎯 Objective

Analyze weekly study hours and discover:

- 📈 Total study time
- ⭐ Most productive day
- 📉 Least productive day
- 📊 Average daily study hours
- 🎯 Weekly goal progress

##  Tools Used

| Tool | Purpose |
|--------|---------|
| Python | Programming |
| NumPy | Data Analysis |
| Jupyter Notebook | Development Environment |

##  Weekly Study Data

| Day | Hours |
|------|------:|
| Monday | 2 |
| Tuesday | 3 |
| Wednesday | 1 |
| Thursday | 4 |
| Friday | 2 |
| Saturday | 5 |
| Sunday | 3 |


## 🔥 Key Insights

✅ Total Study Hours Calculated

✅ Most Productive Day Identified

✅ Average Study Time Computed

✅ Goal Progress Tracked


###  Learning Outcome

This project helped me practice:

- NumPy Arrays
- Statistical Analysis
- Data Aggregation
- Productivity Tracking
```

```python
import numpy as np
```

```python
days = np.array([
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
    "Sunday"
])

hours = np.array([
    2,
    3,
    1,
    4,
    2,
    5,
    3
])
```

```python
total_hours = np.sum(hours)


print("Total Study Hours:", total_hours)
```

```python
best_day = days[np.argmax(hours)]
least_day = days[np.argmin(hours)]

print("Most Productive Day:", best_day)
print("Least Productive Day:", least_day)
```

```python
average_hours = np.mean(hours)

print("Average Daily Hours:", round(average_hours, 2))
```

```python
print("\n========== AI STUDY DASHBOARD ==========")

print("Total Study Hours:", total_hours)

print("Average Daily Hours:", round(average_hours,2))

print("Most Productive Day:", best_day)

print("Least Productive Day:", least_day)
```

```python
study_percentage = (hours / total_hours) * 100

print("\nStudy Time Distribution")

for i in range(len(days)):
    print(days[i], ":", round(study_percentage[i],2), "%")
```

```python
goal = 30

remaining_hours = goal - total_hours

print("Hours Needed To Reach Weekly Goal:", remaining_hours)
```
